```python
import gym
import torch
from torch import optim

%load_ext autoreload
%autoreload 2

import common
```

```python
env = gym.make('CartPole-v1')
env.seed(1); torch.manual_seed(1);

policy = common.MLPPolicy(env)
optimizer = optim.Adam(policy.parameters(), lr=0.01)
updater = common.PGUpdater(optimizer=optimizer, gamma=.99)

common.train_loop(
    env=env,
    policy=policy,
    updater=updater,
    n_eposodes=1000)
```

```python
_, _, _, frames = common.play_episode(env, policy, render=True)

import moviepy.editor as mpy

clip = mpy.ImageSequenceClip(frames, fps=24)
mpy.ipython_display(clip, center=False, width=600)
```
