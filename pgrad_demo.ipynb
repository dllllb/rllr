```python
import gym
import torch
from torch import optim
%matplotlib inline

%load_ext autoreload
%autoreload 2

import pgrad
```

```python
env = gym.make('CartPole-v1')
env.seed(1)
torch.manual_seed(1)

nn = pgrad.MLPPolicy(env)
optimizer = optim.Adam(nn.parameters(), lr=0.01)
updater = pgrad.PGUpdater(optimizer, gamma=.99)
policy = pgrad.NNPolicy(pgrad.MLPPolicy(env), updater)

pgrad.train_loop(env=env, policy=policy, n_eposodes=1000)
```

```python
time, frames = pgrad.play_episode(env, policy, render=True)

import moviepy.editor as mpy

clip = mpy.ImageSequenceClip(frames, fps=24)
mpy.ipython_display(clip, center=False, width=600)
```
